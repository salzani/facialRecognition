```python
import cv2
import dlib
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img/px-people.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Ensure the path to the Haar cascade XML file is correct
classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

faces_detected = classifier.detectMultiScale(img_gray, 1.3, 5)

len(faces_detected)
```

```python
faces_detected[0]
```

```python
img_with_detections = img.copy()

for (x, y, w, h) in faces_detected:
    cv2.rectangle(img_with_detections, (x, y), (x+w, y+h), (255, 0, 0), 3)

plt.imshow(img_with_detections)
```

### **Slice notation**

```python
import os
image_face = 0

for (x, y, w, h) in faces_detected:
    image_face += 1
    image_roi = img[y:y+h, x:x+w]
    image_roi = cv2.cvtColor(image_roi, cv2.COLOR_RGB2BGR)
    cv2.imwrite(os.path.join("faces", "face_" + str(image_face) + ".png"), image_roi)
```
